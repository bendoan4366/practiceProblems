```python
#Weather data practice
weatherdf = pandas.read_excel(r"assets/verlegenhuken.xlsx")
weatherdf
```

```python
weatherdf["Date"] = weatherdf.apply(lambda row: dt(row['Year'], row['Month'], row['Day']), axis=1)

date = weatherdf["Date"]
temperature = weatherdf["Temperature"]
pressure = weatherdf["Pressure"]

weatherf = figure(plot_width=1000,plot_height=800, toolbar_location='below')
weatherf.title.text="Weather and Shit"
weatherf.title.text_color="Gray"
weatherf.title.text_font_style="bold"
weatherf.xaxis.axis_label="Date"
weatherf.yaxis.axis_label="Temperature" 

weatherf.line(date, temperature, color = "blue", alpha=.25)
```
