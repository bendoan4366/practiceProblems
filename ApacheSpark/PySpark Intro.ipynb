```python
import findspark
findspark.init()

import pyspark as ps
from pyspark.sql import SparkSession
```

```python
spark = ps.sql.SparkSession.builder.master("local[4]").appName("spark-intro").getOrCreate()

sc = spark.sparkContext
```

```python
import random

n = 100

heads = (sc.parallelize(range(n))
        .map(lambda _: random.random())
        .filter(lambda r: r <=0.5)
        .count())

tails = n - heads
ratio = 1. * heads/n

print("Heads " + str(heads))
print("Tails " + str(tails))
print("ratio " + str(ratio))
```

```python
rdd= sc.parallelize(range(n))
rdd2 = rdd.map(lambda x: random.random())
rdd2.collect()
```

```python
def is_prime(number):
    factor_min = 2
    factor_max = int(number ** 0.5) + 1
    for factor in range(factor_min, factor_max):
        if number % factor == 0:
            return False
    return True
```

```python
numbers = range(2, 100)

primes = (sc.parallelize(numbers)
    .filter(is_prime)
    .collect())

print(primes)
```
